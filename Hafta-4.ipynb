```python
import pandas as pd
```

```python
emp = pd.read_csv("./resources/employees.csv"  ,parse_dates = ["Start Date"])
emp.info()
# parse_dates kullanarak start date sütununun içeriğini tarih olarak dönüştürdük
# önceden object idi.
```

```python
emp["Last Login Time"] = pd.to_datetime(emp["Last Login Time"])
emp.info()
# pd.to_datetime fonksiyonu bir seriyi tarih ve saat nesnesine dönüştürür.
```

```python
emp.head() # yukardan ilk 5 listeyi getirir.
```

```python
emp["Senior Management"] = emp["Senior Management"].astype(bool) 
emp.info()
# astype burada istenilen sütunun veri tipini değiştirmede kullanılır.
```

```python
emp.Gender.unique()
# gender sütunundaki benzersiz değerleri döndürür.
```

```python
emp.Gender = emp.Gender.astype("category")
emp.info()
# burada Gender sütunundaki değerlerin veri tipini "category" olarak değiştirir
```

```python
emp[emp["First Name"].duplicated()].sort_values(by="First Name")
#  emp["First Name"].duplicated(): Bu ifade, "First Name" sütunundaki her bir
# değerin tekrar edilip edilmediğini kontrol eder.
# .sort_values(by="First Name"): Bu metot, belirtilen sütuna göre DataFrame'i
# sıralar. Bu durumda, "First Name" sütununa göre sıralama yapılır.
```
